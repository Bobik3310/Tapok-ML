```python
#  random seed  
np.random.seed(1)

#    
X1 = np.random.uniform(0, 5, 100)  # 100    0  5
X2 = np.random.uniform(0, 5, 100)  #  100   0  5

#    
X = np.hstack((X1[:, None], X2[:, None]))
# : X.shape = (100, 2)
# X[:, 0] = X1, X[:, 1] = X2

#    ( )
Y = np.where(X1 + X2 < 5, 0, 1)
#  0:  X1 + X2 < 5
#  1:  X1 + X2 >= 5
```

```python
#   LogisticRegression  sklearn
from sklearn.linear_model import LogisticRegression

#      L2-
clf = LogisticRegression(penalty='l2')

#     X ()  Y ( )
clf.fit(X, Y)
```

```python
from matplotlib.colors import ListedColormap

def plot_separating_surface(X, y, cls, view_support=False):
    #     
    x_min = min(X[:, 0]) - 0.1  #  X   
    x_max = max(X[:, 0]) + 0.1  #  X  
    y_min = min(X[:, 1]) - 0.1  #  Y  
    y_max = max(X[:, 1]) + 0.1  #  Y  
    h = 0.005  #   ()
    
    #  
    cm = plt.cm.RdBu  # -  
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])  #    
    
    #   
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    #     
    Z = cls.predict(np.c_[xx.ravel(), yy.ravel()])
    
    #  
    plt.figure(figsize=(8, 8))
    
    #    
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=40, cmap=cm_bright)
    
    #     ( SVM)
    if view_support:
        plt.scatter(X[cls.support_, 0], X[cls.support_, 1],
                    c=y[cls.support_], edgecolors='k', s=150, cmap=cm_bright)
    
    #     
    Z = Z.reshape(xx.shape)
    
    #    
    plt.contourf(xx, yy, Z, cmap=cm, alpha=.3)
    
    #  
    plt.title(" ", size=15)
    plt.xlabel(r'$x_1$', size=15)  #   x
    plt.ylabel(r'$x_2$', size=15)  #   x
```

```python
plot_separating_surface(X, Y, clf)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score
import numpy as np

#    
base_logloss = 0.3498
base_auc = 0.8721

print(" :")
print(f"Test logloss = {base_logloss:.4f}")
print(f"Test roc auc score = {base_auc:.4f}")
print("-" * 50)

#    
def evaluate_model(pipeline, X_train, y_train, X_test, y_test, description):
    model = pipeline.fit(X_train.drop(columns=["Date"]), y_train)
    y_pred = model.predict_proba(X_test.drop(columns=["Date"]))[:, 1]
    logloss = log_loss(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)

    print(f"{description}:")
    print(f"Test logloss = {logloss:.4f} ({'' if logloss < base_logloss else ''} {abs(logloss - base_logloss):.4f})")
    print(f"Test roc auc score = {auc:.4f} ({'' if auc > base_auc else ''} {abs(auc - base_auc):.4f})")
    print("-" * 30)

    return logloss, auc


#       
best_pipeline = Pipeline(steps=[
    ('preprocessor', column_transformer),
    ('classification', LogisticRegression(
        solver='lbfgs',
        max_iter=1000,
        C=0.5,  #  
        class_weight='balanced'  #   
    ))
])
logloss_best, auc_best = evaluate_model(best_pipeline, X_train, y_train, X_test, y_test,
                                      "  ")

#   
results = {
    " ": (logloss_best, auc_best)
}
```
