```python
#     

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
```

```python
#   
import time
import numpy as np
import matplotlib.pyplot as plt

#    (,    )
from unittests import research_functions
from unittests import research_functions_vectorised

#  
plt.style.use('seaborn-v0_8-darkgrid')

#     
def plot_comparison(sizes, times_non_vec, times_vec, func_name, yscale_linear=True):
    """
       .
    
    Args:
        sizes:  
        times_non_vec:   
        times_vec:   
        func_name:  
        yscale_linear:    (True)   (False)
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    #       
    times_non_vec_ms = [t * 1000 for t in times_non_vec]
    times_vec_ms = [t * 1000 for t in times_vec]
    
    #  1:  
    ax1.plot(sizes, times_non_vec_ms, 'r-o', linewidth=2, markersize=6, label='')
    ax1.plot(sizes, times_vec_ms, 'g-s', linewidth=2, markersize=6, label='')
    ax1.set_xlabel('  (n)', fontsize=12)
    ax1.set_ylabel('  ()', fontsize=12)
    ax1.set_title(f'{func_name}:   ', fontsize=13, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.legend(fontsize=11)
    
    if not yscale_linear:
        ax1.set_yscale('log')
        ax1.set_ylabel('  (, log scale)', fontsize=12)
    
    #  2: 
    speedup = np.array(times_non_vec) / np.array(times_vec)
    ax2.plot(sizes, speedup, 'b-^', linewidth=2, markersize=8)
    ax2.axhline(y=1, color='r', linestyle='--', alpha=0.5, label=' ')
    ax2.fill_between(sizes, 1, speedup, alpha=0.2, color='green')
    ax2.set_xlabel('  (n)', fontsize=12)
    ax2.set_ylabel(' ()', fontsize=12)
    ax2.set_title(f'{func_name}:   ', fontsize=13, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    ax2.legend(fontsize=11)
    
    #     
    max_speedup = max(speedup)
    max_idx = np.argmax(speedup)
    ax2.annotate(f': {max_speedup:.1f}x\n n={sizes[max_idx]}',
                xy=(sizes[max_idx], max_speedup),
                xytext=(sizes[max_idx], max_speedup * 1.1),
                arrowprops=dict(arrowstyle='->', color='black'),
                fontsize=10,
                ha='center')
    
    plt.tight_layout()
    plt.show()
    
    return speedup

#  
print("="*70)
print("  ")
print("="*70)

are_multisets_equal_data = []
max_prod_mod_3_data = []
convert_image_data = []
rle_scalar_data = []
cosine_distance_data = []

sizes = {
    'are_multisets_equal': [10, 100, 1000, 10000],
    'max_prod_mod_3': [10, 100, 1000, 10000],
    'convert_image': [10, 100, 1000],
    'rle_scalar': [10, 100, 1000, 10000],
    'cosine_distance': [10, 100]
}

# 1. are_multisets_equal
print("   are_multisets_equal...")
for n in sizes['are_multisets_equal']:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data.append((x, y))

# 2. max_prod_mod_3
print("   max_prod_mod_3...")
for n in sizes['max_prod_mod_3']:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data.append(x)

# 3. convert_image
print("   convert_image...")
for n in sizes['convert_image']:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data.append((image, weights))

# 4. rle_scalar
print("   rle_scalar...")
for n in sizes['rle_scalar']:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes
    rle_scalar_data.append((x, y))

# 5. cosine_distance
print("   cosine_distance...")
for n in sizes['cosine_distance']:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data.append((X, Y))

print("  !\n")

#     
results = {}

# 1. max_prod_mod_3
print("="*70)
print("1. : max_prod_mod_3")
print("="*70)

max_prod_mod_3_time = []
max_prod_mod_3_vectorised_time = []

for i, x in enumerate(max_prod_mod_3_data):
    n = sizes['max_prod_mod_3'][i]
    
    #   (  )
    time_non_vec, result_non_vec = count_time(research_functions.max_prod_mod_3, x.tolist())
    max_prod_mod_3_time.append(time_non_vec)
    
    #   (  numpy )
    time_vec, result_vec = count_time(research_functions_vectorised.max_prod_mod_3, x)
    max_prod_mod_3_vectorised_time.append(time_vec)
    
    #   
    if not np.isclose(result_non_vec, result_vec):
        print(f"  :    n={n}: {result_non_vec} != {result_vec}")
    
    print(f"  n={n:6d}: ={time_non_vec*1000:7.3f} , "
          f"={time_vec*1000:7.3f} , "
          f"={time_non_vec/time_vec:6.1f}x")

#  
speedup_max_prod = plot_comparison(
    sizes['max_prod_mod_3'],
    max_prod_mod_3_time,
    max_prod_mod_3_vectorised_time,
    "max_prod_mod_3",
    yscale_linear=False
)

results['max_prod_mod_3'] = {
    'sizes': sizes['max_prod_mod_3'],
    'times_non_vec': max_prod_mod_3_time,
    'times_vec': max_prod_mod_3_vectorised_time,
    'speedup': speedup_max_prod
}

# 2. are_multisets_equal
print("\n" + "="*70)
print("2. : are_multisets_equal")
print("="*70)

are_multisets_equal_time = []
are_multisets_equal_vectorised_time = []

for i, (x, y) in enumerate(are_multisets_equal_data):
    n = sizes['are_multisets_equal'][i]
    
    #  
    time_non_vec, result_non_vec = count_time(research_functions.are_multisets_equal, 
                                               x.tolist(), y.tolist())
    are_multisets_equal_time.append(time_non_vec)
    
    #  
    time_vec, result_vec = count_time(research_functions_vectorised.are_multisets_equal, x, y)
    are_multisets_equal_vectorised_time.append(time_vec)
    
    #  
    if result_non_vec != result_vec:
        print(f"  :    n={n}: {result_non_vec} != {result_vec}")
    
    print(f"  n={n:6d}: ={time_non_vec*1000:7.3f} , "
          f"={time_vec*1000:7.3f} , "
          f"={time_non_vec/time_vec:6.1f}x")

speedup_multisets = plot_comparison(
    sizes['are_multisets_equal'],
    are_multisets_equal_time,
    are_multisets_equal_vectorised_time,
    "are_multisets_equal"
)

results['are_multisets_equal'] = {
    'sizes': sizes['are_multisets_equal'],
    'times_non_vec': are_multisets_equal_time,
    'times_vec': are_multisets_equal_vectorised_time,
    'speedup': speedup_multisets
}

# 3. convert_image
print("\n" + "="*70)
print("3. : convert_image")
print("="*70)

convert_image_time = []
convert_image_vectorised_time = []

for i, (image, weights) in enumerate(convert_image_data):
    n = sizes['convert_image'][i]
    
    #  
    time_non_vec, result_non_vec = count_time(research_functions.convert_image, 
                                               image.tolist(), weights.tolist())
    convert_image_time.append(time_non_vec)
    
    #  
    time_vec, result_vec = count_time(research_functions_vectorised.convert_image, image, weights)
    convert_image_vectorised_time.append(time_vec)
    
    #  
    if not np.allclose(result_non_vec, result_vec, rtol=1e-10):
        print(f"  :    n={n}")
    
    print(f"  n={n:6d}: ={time_non_vec*1000:7.3f} , "
          f"={time_vec*1000:7.3f} , "
          f"={time_non_vec/time_vec:6.1f}x")

speedup_convert = plot_comparison(
    sizes['convert_image'],
    convert_image_time,
    convert_image_vectorised_time,
    "convert_image",
    yscale_linear=False
)

results['convert_image'] = {
    'sizes': sizes['convert_image'],
    'times_non_vec': convert_image_time,
    'times_vec': convert_image_vectorised_time,
    'speedup': speedup_convert
}

# 4. rle_scalar
print("\n" + "="*70)
print("4. : rle_scalar")
print("="*70)

rle_scalar_time = []
rle_scalar_vectorised_time = []

for i, (x, y) in enumerate(rle_scalar_data):
    n = sizes['rle_scalar'][i]
    
    #  
    time_non_vec, result_non_vec = count_time(research_functions.rle_scalar, 
                                               x.tolist(), y.tolist())
    rle_scalar_time.append(time_non_vec)
    
    #  
    time_vec, result_vec = count_time(research_functions_vectorised.rle_scalar, x, y)
    rle_scalar_vectorised_time.append(time_vec)
    
    #  
    if not np.allclose(result_non_vec, result_vec, rtol=1e-10):
        print(f"  :    n={n}")
    
    print(f"  n={n:6d}: ={time_non_vec*1000:7.3f} , "
          f"={time_vec*1000:7.3f} , "
          f"={time_non_vec/time_vec:6.1f}x")

speedup_rle = plot_comparison(
    sizes['rle_scalar'],
    rle_scalar_time,
    rle_scalar_vectorised_time,
    "rle_scalar",
    yscale_linear=False
)

results['rle_scalar'] = {
    'sizes': sizes['rle_scalar'],
    'times_non_vec': rle_scalar_time,
    'times_vec': rle_scalar_vectorised_time,
    'speedup': speedup_rle
}

# 5. cosine_distance
print("\n" + "="*70)
print("5. : cosine_distance")
print("="*70)

cosine_distance_time = []
cosine_distance_vectorised_time = []

for i, (X, Y) in enumerate(cosine_distance_data):
    n = sizes['cosine_distance'][i]
    
    #  
    time_non_vec, result_non_vec = count_time(research_functions.cosine_distance, 
                                               X.tolist(), Y.tolist())
    cosine_distance_time.append(time_non_vec)
    
    #  
    time_vec, result_vec = count_time(research_functions_vectorised.cosine_distance, X, Y)
    cosine_distance_vectorised_time.append(time_vec)
    
    #  
    if not np.allclose(result_non_vec, result_vec, rtol=1e-10):
        print(f"  :    n={n}")
    
    print(f"  n={n:6d}: ={time_non_vec*1000:7.3f} , "
          f"={time_vec*1000:7.3f} , "
          f"={time_non_vec/time_vec:6.1f}x")

speedup_cosine = plot_comparison(
    sizes['cosine_distance'],
    cosine_distance_time,
    cosine_distance_vectorised_time,
    "cosine_distance"
)

results['cosine_distance'] = {
    'sizes': sizes['cosine_distance'],
    'times_non_vec': cosine_distance_time,
    'times_vec': cosine_distance_vectorised_time,
    'speedup': speedup_cosine
}

#  
print("\n" + "="*70)
print(" :  ")
print("="*70)

#   
print(f"{'':<25} {' ':<15} {'. ()':<15} {'. ()':<15} {'':<10}")
print("-" * 85)

for func_name, data in results.items():
    sizes_list = data['sizes']
    times_non_vec = data['times_non_vec']
    times_vec = data['times_vec']
    speedup = data['speedup']
    
    #     
    for i in range(len(sizes_list)):
        n = sizes_list[i]
        time_non_vec_ms = times_non_vec[i] * 1000
        time_vec_ms = times_vec[i] * 1000
        speedup_val = speedup[i]
        
        #  
        func_display = func_name if i == 0 else ""
        print(f"{func_display:<25} {n:<15} {time_non_vec_ms:<15.3f} {time_vec_ms:<15.3f} {speedup_val:<10.1f}x")

print("\n" + "="*70)
print("   ")
print("="*70)

#  
print("\n1.  :")
print("   -     ")
print("   -    5x  1000x    ")

print("\n2.   :")
print(f"   a) max_prod_mod_3:")
print(f"      *  : {max(speedup_max_prod):.1f}x  n={sizes['max_prod_mod_3'][np.argmax(speedup_max_prod)]}")
print(f"      * : NumPy      ")

print(f"   b) are_multisets_equal:")
print(f"      *  : {max(speedup_multisets):.1f}x  n={sizes['are_multisets_equal'][np.argmax(speedup_multisets)]}")
print(f"      * :       NumPy")

print(f"   c) convert_image:")
print(f"      *  : {max(speedup_convert):.1f}x  n={sizes['convert_image'][np.argmax(speedup_convert)]}")
print(f"      * :       broadcast")

print(f"   d) rle_scalar:")
print(f"      *  : {max(speedup_rle):.1f}x  n={sizes['rle_scalar'][np.argmax(speedup_rle)]}")
print(f"      * :     ")

print(f"   e) cosine_distance:")
print(f"      *  : {max(speedup_cosine):.1f}x  n={sizes['cosine_distance'][np.argmax(speedup_cosine)]}")
print(f"      * :      NumPy")

print("\n3.    :")
print("   -         ")
print("   - :         ")
print("   -    n    -  ")

print("\n4.    NumPy:")
print("   -   C-   Python")
print("   -    (SIMD: SSE, AVX)")
print("   -   -")
print("   -       Python")
print("   -     (BLAS/LAPACK)")

print("\n5.  :")
print("   -        ")
print("   -       ")
print("   -        ")
print("   -   Python    NumPy ")

#      
print("\n" + "="*70)
print(" :     ")
print("="*70)

#     
func_names = list(results.keys())
max_speedups = [max(data['speedup']) for data in results.values()]
avg_speedups = [np.mean(data['speedup']) for data in results.values()]

fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(func_names))
width = 0.35

bars1 = ax.bar(x - width/2, max_speedups, width, label=' ', color='skyblue', edgecolor='black')
bars2 = ax.bar(x + width/2, avg_speedups, width, label=' ', color='lightcoral', edgecolor='black')

ax.set_xlabel('', fontsize=12)
ax.set_ylabel(' ()', fontsize=12)
ax.set_title('      ', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(func_names, rotation=45, ha='right')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis='y')

#    
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}x',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=9)

plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("")
print("="*70)
print("   NumPy   ")
print("    .  ")
print("       .")
```
